I have a bank that customers can open a bank account online using my web application.

clients from some countries are not allowed to open an account.

```python
from time import sleep
country_ips = [
    { "ip": "2.0.0.0", "country": "Germany" },
    { "ip": "1.0.0.0", "country": "China" },
    { "ip": "3.0.0.0", "country": "United States" },
    { "ip": "5.0.0.0", "country": "United Kingdom" },
    { "ip": "31.0.0.0", "country": "France" },
    { "ip": "39.0.0.0", "country": "India" },
    { "ip": "45.0.0.0", "country": "Ghana" },  # Corrected spelling
    { "ip": "46.1.14.17", "country": "North Korea" }  # Removed leading zeros
]
forbidden_countries = ["China", "North Korea"]

def check_ip(ip):
    """
    Check if the IP address belongs to a forbidden country.
    """
    for ip_info in country_ips:
        # print("The IP is processing to check the source country...")
        if ip == ip_info['ip']:
            country_x = ip_info['country']
            print(f"the country is {ip_info['country']}")
            print("System is checking the country...")
            sleep(2)
            if country_x in forbidden_countries:
                print(f"⛔️ Access denied from {country_x}") 
            else:
                print(f"✅ Access to signup page to the user from {country_x}!") 
            
        

check_ip("46.1.14.17")
```

```python
cargo = [
    {
        "port_of_loading": "Montreal",
        "port_of_discharge": "roterdam",
        "destination": "Zaandam",

    }
]

shipping_modes = [ "Barge", "Truck" , "Barge&Truck"]

container_size =  ["20ft", "40hc", "40ft"]

def calculate_shipping_cost(request):
    base_rates = {
        ("Barge", "20ft"): 75,
        ("Truck", "20ft"): 250,
        ("Barge&Truck", "20ft"): 130
    }

    # Adjustments by container size
    size_multipliers = {
        "20ft": 1.00,
        "40ft": 1.25,
        "40hc": 1.40
    }

    cost_results = []

    for shipment in request:
        mode = shipment["Shipping_mode"]
        size = shipment["Container_size"]

        base_rate = base_rates.get((mode, "20ft"), 0)  # Use 20ft as base, then apply multiplier
        multiplier = size_multipliers.get(size, 1.0)

        final_cost = base_rate * multiplier

        cost_results.append({
            "port_of_loading": shipment["port_of_loading"],
            "port_of_discharge": shipment["port_of_discharge"],
            "destination": shipment["destination"],
            "Shipping_mode": mode,
            "Container_size": size,
            "cost": round(final_cost, 2)
        })

    return cost_results

cargo = [
    {
        "port_of_loading": "Montreal",
        "port_of_discharge": "Rotterdam",
        "destination": "Zadaam",
        "Shipping_mode": "Barge&Truck",
        "Container_size": "40ft"
    }
]

calculate_shipping_cost(cargo)



```
