# error handling using `TRY`

The try block lets you test a block of code for errors.
It’s often used when you expect something might go wrong — like reading a file, dividing by zero, or accessing a key in a dictionary.

```python
my_numb = input("Enter a number: ")

print(my_numb + 10)
```

```python
# Catching division by zero
# try:
#     x = 10 / 0
#     print(x)
# except ZeroDivisionError:
#     print("You can't divide by zero!")


def my_fun(x):
    try:
        print(10 / x)
    except ZeroDivisionError:
        print("You can't divide by zero!")


my_fun(2.5)
```

```python
# Reading a non-existent file * * * * *
try:
    with open("nonexistent.txt", "r") as f:
        content = f.read()
except FileNotFoundError:
    print("File not found!")
except NameError:
    print("Variable not defined!")
```

```python
# try-except-else-finally


# else block:
	# •	Runs only if no exceptions were raised in the try block.
#  finally block:
    # •	Runs regardless of whether an exception was raised or not.
try:
    x = 5 / 1
except ZeroDivisionError:
    print("Error: Division by zero.")
else:
    print("Everything went fine!")
finally:
    print("This always runs.")
```

```python
# try on syntax error


a = "5"
b = 993

try:
    print(a + b)
except TypeError:
    print(int(a) + int(b))
```

```python
# try on NameError
userss = "hermann"

try:
    print(userss)
except NameError:
    print(">>> Variable 'userss' is not defined. Please check your code.")
else:
    print(">>> all good here")
finally:
    print(">>> This block always executes, regardless of errors.")

```

```python
def safe_divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("You can't divide by zero. Let's fix that.")
        y = float(input("Please enter a non-zero number for y: "))
        result = x / y  # retry division
    return result

a = 10
b = 0  # This will trigger exception

print("Result:", safe_divide(a, b))
```

```python
import requests
from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/vehicles')
def get_vehicles():
    api_url = "https://68376b322c55e01d1849cb71.mockapi.io/api/v1/vehicles"
    response = requests.get(api_url)
    if response.status_code == 200:
        return jsonify(response.json())
    else:
        return jsonify({"error": response.status_code, "message": response.text}), response.status_code

if __name__ == '__main__':
    app.run(debug=True)
```

```python
import requests
import time

books = []
genres = ["fiction", "science", "history", "fantasy", "romance"]

for genre in genres:
    for start in range(0, 120, 40):  # 3 pages per genre, 40 per page
        url = f"https://www.googleapis.com/books/v1/volumes"
        params = {
            "q": f"subject:{genre}",
            "startIndex": start,
            "maxResults": 40,
        }
        response = requests.get(url, params=params)
        data = response.json()

        for item in data.get("items", []):
            title = item["volumeInfo"].get("title", "No Title")
            authors = item["volumeInfo"].get("authors", ["Unknown"])
            books.append({
                "title": title,
                "authors": authors,
                "genre": genre
            })

        time.sleep(0.5)  # Avoid hitting rate limits

# Show first 10 results
for book in books[:10]:
    print(book)

print(f"\nTotal books fetched: {len(books)}")
```
