# `JSON` (JavaScript Object Notation) 

is a text-based data format that is commonly used for transmitting data between a server and a web application. <br>
It is a lightweight and human-readable format that's also easy for machines to parse. Despite its name, JSON is language-independent and can be used with various programming languages and platforms. 




### important signs:
- [] -> entire documents
- {} -> entire object
- : -> key and values
- "" -> keys
- , -> divider

### write in a file context using `open(<file name>, "w")` from a list
